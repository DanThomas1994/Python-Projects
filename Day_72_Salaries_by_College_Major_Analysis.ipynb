```python
import pandas as pd
# Only the last line in the cell gets printed out automatically; to print out lines above it, a print statement is required.
df = pd.read_csv('salaries_by_college_major.csv') # Create a data frame by reading the csv.
# print(df)
# print(df.head()) # Print the 'head' of the dataset (the first 5 rows).
# print(df.shape) # Print the number of rows and columns.
# df.columns
df.isna() # Find any cells which contain NaN.
# print(df.tail)

# Create a new dataframe, removing the row containing the NaN values:
cleaned_df = df.dropna() # Remove the row containing the NaN values using the dropna() method.
# cleaned_df

# Access a particular column in the dataframe:
cleaned_df['Starting Median Salary']

# Access the max value in this column:
cleaned_df['Starting Median Salary'].max() # Find the max value.
cleaned_df['Starting Median Salary'].idxmax() # Find the row corresponding to the max value.
cleaned_df['Undergraduate Major'].loc[43] # Find the major corresponding to the row with the max value.

# Achieve same result with a different method:
cleaned_df['Undergraduate Major'][43]

# Retrieve an entire row:
cleaned_df.loc[43]

# Find the highest mid-career earnings:
cleaned_df['Mid-Career Median Salary'].max()
f"Index for the max mid-career salary: {cleaned_df['Mid-Career Median Salary'].idxmax()}"
cleaned_df['Undergraduate Major'][8]

# Major with the lowest starting salary:
cleaned_df['Starting Median Salary'].min()
cleaned_df['Undergraduate Major'].loc[cleaned_df['Starting Median Salary'].idxmin()]

# Major with the lowest mid-career salary:
cleaned_df['Mid-Career Median Salary'].min()
cleaned_df['Undergraduate Major'].loc[cleaned_df['Mid-Career Median Salary'].idxmin()]

spread_column = cleaned_df['Mid-Career 90th Percentile Salary'].subtract(cleaned_df['Mid-Career 10th Percentile Salary'])
cleaned_df.insert(1, 'Spread', spread_column) # 1 is the position of the column.
cleaned_df.head()

low_risk = cleaned_df.sort_values('Spread',ascending=True) # Lowest risk here means the smallest difference between
# the mid-career 10th and 90th percentiles.
low_risk[['Undergraduate Major', 'Spread']].head()

highest_potential = cleaned_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()

greatest_spread = cleaned_df.sort_values('Spread', ascending=False) # Sort the rows in descending order according to the
# values in the spread column.
greatest_spread[['Undergraduate Major', 'Spread']].head()

# Using the groupby() method:
cleaned_df.groupby('Group').count() # Count how many majors we have in each category.
pd.options.display.float_format = '{:,.2f}'.format # Print the numbers in a format like 1,012.45.
cleaned_df.groupby('Group').mean(numeric_only=True) # Find the average salary by group. numeric_only = True is needed for
# this line of code to work.
# cleaned_df.head()
```
